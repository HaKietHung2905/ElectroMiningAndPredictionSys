```python
columns = ['HourDK', 'Consumption_MWh']

df_MA = df[(df['DK36Code'] == 'PR') & (df['season'] == 'Summer')][columns]
df_MA['DateTime'] = pd.to_datetime(df_MA['HourDK'])

df_MA = df_MA[(df_MA['DateTime'].dt.year == 2023)][columns]

df_MA
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import skfuzzy as fuzz

# Assuming df_MA is your DataFrame containing 'HourDK' and 'Consumption_MWh' columns
columns = ['HourDK', 'Consumption_MWh']
data = df_MA[columns].copy()  # Use a copy to avoid SettingWithCopyWarning

# Convert 'HourDK' to numerical representation (Unix timestamp)
data['HourDK'] = pd.to_datetime(data['HourDK']).dt.hour

# Ensure 'Consumption_MWh' is of float type
data['Consumption_MWh'] = data['Consumption_MWh'].astype(float)

# Convert DataFrame to numpy array
alldata = data.to_numpy().T  # Transpose the data for cmeans

# Define cluster colors
colors = ['b', 'orange', 'g', 'r', 'c', 'm', 'y', 'k', 'Brown', 'ForestGreen']
```

```python
data['HourDK']
```

```python

# Visualize the data
fig0, ax0 = plt.subplots()
ax0.scatter(data['HourDK'], data['Consumption_MWh'], color='gray')
ax0.set_title('Original Data')
ax0.set_xlabel('HourDK')
ax0.set_ylabel('Consumption_MWh')

```

```python
# Perform fuzzy clustering and visualize

fig1, axes1 = plt.subplots(1, 2, figsize=(10, 10))
fpcs = []

ncenters = 3  # Fixed number of clusters
axes1.flatten()
```

```python
fig1, axes1 = plt.subplots(1, 2, figsize=(10, 10))
fpcs = []
ncenters = 4  # Fixed number of clusters

# Iterate over axes1 properly
for ax in axes1.flatten():
    cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
        alldata, ncenters, 2, error=0.005, maxiter=1000, init=None)

    # Store fpc values for later
    fpcs.append(fpc)

    # Plot assigned clusters
    cluster_membership = np.argmax(u, axis=0)
    for j in range(ncenters):
        ax.scatter(df_MA['HourDK'][cluster_membership == j],
                   df_MA['Consumption_MWh'][cluster_membership == j],
                   color=colors[j], label=f'Cluster {j+1}')

    # Mark the center of each fuzzy cluster
    for pt in cntr:
        ax.plot(pt[0], pt[1], 'rs', markersize=10)

    ax.set_title(f'Centers = {ncenters}; FPC = {fpc:.2f}')
    ax.set_xlabel('HourDK (scaled)')
    ax.set_ylabel('Consumption_MWh (scaled)')
    ax.legend()

fig1.tight_layout()
plt.show()
```
